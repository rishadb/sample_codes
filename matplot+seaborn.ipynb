```python
##Bar chart from csv
plt.clf(); from collections import Counter; import pandas as pd; 
csv_reader = pd.read_csv("/content/drive/MyDrive/DATA/matplotdata.csv")
counter = Counter() #to count elements in list passed as dicts
lang_worked = csv_reader["LanguagesWorkedWith"]
for row in lang_worked:
  counter.update(row.split(";"))
langs = []; popu= []
for item in counter.most_common(15):
  langs.append(item[0])
  popu.append(item[1])
langs.reverse(); popu.reverse() #to sort high to low
plt.barh(langs, popu); plt.show()
```

```python
#pie
plt.pie([1,2,3], labels = ["a", "b", "c"]); plt.show
plt.pie(popu, labels = langs, autopct = '%1.1f%%'); plt.show
```
