```python
df = pd.read_csv("/content/drive/MyDrive/DATA/titanic.csv")
df.head(); df.tail(6); df.info(); df.describe().T; 
#18 above male survived count
df.Age[(df.Sex=="male") & (df.Age>=18) ]; df.PassengerId[df.Sex=="male"].count()
#female and children survived
s= df.Survived[(df.Sex == 'male') & (df.Age > 60)]# == (s == 1).sum()/ s.count()
#transform Age and Fare cols 
df.transform({'Age': np.sqrt,'Fare': np.exp,})#cant apply aggregating funcs like sum, avg...use apply for that, also cant use deff cols in same operation
#see defference of 2 cols
def sub_cols(x):
   return x["Age"] -x["Fare"]
df.apply(sub_cols, axis = 1)
#median of survived and Fare of all kind of (sexes and ages) combination
df.groupby(["Sex","Age"]).mean(); df.groupby(["Sex", "Age"])["Survived", "Fare"].median()
#median, min, max of survived and Fare of all kind of (sexes and ages) combination
df.groupby(["Sex","Age"]).mean(); df.groupby(["Sex", "Age"])["Survived", "Fare"].aggregate(['min', np.median, max]) #can specify agg funcs as shown here
#median, min, max of specific cols of all kind of (sexes and ages) combination
df.groupby(["Sex","Age"]).mean(); df.groupby(["Sex", "Age"]).aggregate({"Survived":'min', "Fare":np.median,"Fare": max}) #can specify agg funcs as shown here
#filter in all rows of orig table where the groupby filter func is satisfied
def filter_func(x): # the groupby df will be passed in
  return x["Fare"].std() > 4
df.groupby("Sex").filter(filter_func)
#transform dataframe rows with func
df.groupby("Sex").transform(lambda x: x-x.mean()) #x is the same val, but x.mean is groupby vals mean
#normalize Age cols (since agg func and 2 cols used, use apply),Fare sum is the sum of sex group Fares
def norm_data(x):
  x["Age"] /=x["Fare"].sum()
  return x
df.groupby("Sex").apply(norm_data)
#groupby custom way: define 
a = np.ones(100); a = list(a); b = np.zeros(791); b = list(b); a.extend(b); len(a) == len(df)
df.groupby(a).sum()
#set inxdex col 
df2 = df.set_index("Name")
df2.groupby(str.lower()).mean() #here the index is the groupby arg and the index is converted 
df.sort_values(["Pclass", "Sex"], ascending = [True, False])
df.drop(columns=["Parch", "Ticket"])
```

```python
df1 = pd.DataFrame({"employee": ["Bob", "Jake", "Lisa", "Sue", "Joy"], "group":["Accounting", "Engineering", "Engineering", "HR", "staff"]})
df2 = pd.DataFrame({"employee":["Lisa", "Bob", "Jake", "Sue"], "hire_date": [2004, 2002, 2012, 2002]})
df4 = pd.DataFrame({"group": ["Accounting", "Engineering", "HR"], "supervisor": ["Carly", "Guido", "Steve"]})
df5 = pd.DataFrame({"groups": ["Accounting", "Engineering", "Accounting", "Engineering"], "skills": ["math", "coding","Linux", "org"]})
#merge tables
df3 = df1.merge(df2, on = "employee") #== pd.merge(df1, df2)
pd.merge(df4, df3)
pd.merge(df1, df5, left_on = "group", right_on = "groups", how = "inner").drop("groups", axis = 1) #deff col names, drops the duplicate col with deff name #hows inner, outer, left, right
df1.set_index("employee", inplace = True); df2.set_index("employee", inplace = True); pd.merge(df1, df2, left_index = True, right_index = True) #merge on index
df1.reset_index(); pd.merge(df1, df2, left_on = "employee", right_index = True)# merge on index and col
pd.merge(df1, df1, on = "employee", suffixes = ["_L", "_R"]) #== pd.concat([df1, df1], axis = 1) # merge 2 tables with same col names
#concat 2 tables
pd.concat([df3, df4], axis =0)#, verify_integrity = True, ignore_index = True ) #stack two tables along axis mentioned
pd.concat([df3, df4], axis =0, verify_integrity = True, ignore_index = True )# verifyintegrity throws error id duplicate indeces found, but we use ingnore here to ignore and set new index as 0,1,2,...
pd.concat([df3, df4], axis =0, keys = ["A", "B"], join = "inner")#indexes are grouped by the keys here, joins is for inner or outer
df1.append(df2) #same as concat
```
